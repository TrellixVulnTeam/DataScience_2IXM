```python
print("\napplication_df", application_df.shape, "\ndisciplines_df", disciplines_df.shape, "\neducationlevel_df", educationlevel_df.shape, "\nfuzucountries_df", fuzucountries_df.shape, "\nindustries_df", industries_df.shape, "\npositions_df", positions_df.shape, "\npositionskills_df", positionskills_df.shape, "\nskills_df", skills_df.shape, "\nusers_df", users_df.shape, "\nuserskills_df", userskills_df.shape)
```

```python
fuzu_application_df = pd.merge(
    pd.merge(
        pd.merge(
            pd.merge(
                pd.merge(
                    pd.merge(application_df,users_df,left_on="user_id", right_on="id", how="left"),
                    pd.merge(pd.merge(positionskills_df,positions_df,left_on="position_id", right_on="id"),skills_df,left_on="skill_id", right_on="id", how="left"),
                    on="position_id", how="left"),
                industries_df,left_on="industry_id", right_on="id", how="left"),
            educationlevel_df,left_on="education_level_id", right_on="id", how="left"),
        disciplines_df,left_on="discipline_id", right_on="id", how="left"),
    fuzucountries_df,left_on="fuzu_country_id", right_on="id", how="left")
fuzu_application_df.sample(10)
```
