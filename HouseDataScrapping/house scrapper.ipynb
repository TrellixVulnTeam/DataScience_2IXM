```python
def scroller_func(driver):
    SCROLL_PAUSE_TIME = 60
    
    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")
#     print(driver.current_url)

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    #     print(driver.current_url)

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            content = driver.page_source
            break
        last_height = new_height
    return content
```

```python
def scrapper(driver):
    
    content = scroller_func(driver)
    # content = driver.page_source
    soup = bs(content)
    # print (soup)
    VillaCare={'prices':[],'description':[],'location':[]} #Dictionary of houses
    for a in soup.findAll('div',href=False, attrs={'class':'md-listing-info'}):
    #     products.append(a)
    #     print(a)
        price=a.find('div', attrs={'class':'md-listing-price'})
        VillaCare['prices'].append(price.get_text().strip())
    #     print("price = ", price.get_text().strip())
        location=a.find('div', attrs={'class':'md-listing-address'})
        VillaCare['location'].append(location.get_text().strip())
    #     print("rating = ", location.get_text().strip())
        try:
            description=a.find('div', attrs={'class':'md-listings-stats2'})
            description = description.get_text().replace('\n'," ").strip()
    #         print("description = ",description)
    #         print("description2 = ", description2.get_text().strip())
        except:
            description = "-"
    #         print("description = ",description)
    #         continue
    #     print("description after = ",description)
    #     description=a.find('div', attrs={'class':'md-listings-stats2'}).get_text().strip()
    #     print("description = ", description.get_text().strip())
        try:
            description2=a.find('div', attrs={'class':'md-listings-stats'})
            description2 = description2.get_text().replace('\n').strip()
    #         print("description2 = ",description2)
    #         print("description2 = ", description2.get_text().strip())
        except:
            description2 = "-"
    #         print("description2 = ",description2)
    #         continue
    #     print("description2 after = ",description2)

        VillaCare['description'].extend([description +" "+ description2])

    return VillaCare
```

# Anza Properties

```python
def anza_scrapper(scrap_txt):
    anza={'prices':[],'description':[],'location':[]} #Dictionary of houses

    for a in scrap_txt:
    #     products.append(a)
#         print(a)
        price=a.find('span', attrs={'class':'list-price'})
        anza['prices'].append(price.get_text().strip())
    #     print("price = ", price.get_text().strip())
        location=a.find('span', attrs={'class':'list-location'})
        anza['location'].append(location.get_text().strip())
    #     print("rating = ", location.get_text().strip())
        description=a.find('span', attrs={'class':'list-bedbath'}).get_text().strip()
        description2=a.find('div', attrs={'class':'contentIntrotext'}).get_text().strip()
        anza['description'].extend([description +" "+ description2])
        

    return anza
```

```python
city = ['nairobi', 'juja', 'mombasa', 'malindi', 'watamu', 'mtwapa', 'kwale', 'lamu', 'kilifi']
categ = ['commercial', 'residential', 'land']
cat = ['31', '7', '32']
typ = ['sale', 'let']
rec = ['askingprice', 'rent']
anza_dict={'prices':[],'description':[],'location':[]}
# LOAD_PAUSE_TIME = 30
driver = wd.Chrome('./chromedriver')
for x in categ:
    for t in typ:
        for c in city:
            for n in range(1,101):
                link = 'http://www.anza.co.ke/' + x + '/search/search-results?page=' + str(n) + '&order=featured&cat=' + cat[categ.index(x)] + '&query=all&jr_citytown=' + c + '&jr_financials' + rec[typ.index(t)] + '=between_1_10000000&jr_financialsavailablefor=' + t
                driver.get(link)
                
#                 time.sleep(LOAD_PAUSE_TIME)
                
                content = driver.page_source
                soup = bs(content)
                scrap_txt = soup.findAll('div',href=False, attrs={'class':'jrListingInfo'})
                if len(scrap_txt) > 0:
                    dict_anza = anza_scrapper(scrap_txt)
                    anza_dict['prices'].extend(dict_anza['prices'])
                    anza_dict['description'].extend(dict_anza['description'])
                    anza_dict['location'].extend(dict_anza['location'])
                else:
                    break

driver.quit()
anza_df = pd.DataFrame(anza_dict)
print(anza_df.sample(10))
anza_df.to_csv(path_or_buf="anza_properties.csv", index_label= "index")
```

```python
def hasscon_scrapper(scrap_txt):
    hasscon={'prices':[],'description':[],'location':[]} #Dictionary of houses

    for a in scrap_txt:
    #     products.append(a)
#         print(a)
        scrap_info = a.findAll('div', attrs={'class':'mr-pr'})
    #     price=a.find('div', attrs={'class':'mr-pr'})
        mix_info = scrap_info[0].get_text().replace('\t','').strip().split('\n',2)
        price_info = mix_info[2].replace('\n','').strip()
        hasscon['prices'].append(mix_info[2].replace('\n','').strip())
        hasscon['location'].append(mix_info[0])
        desc_info = a.find('div', attrs={'class':'mr-pr1'})
        
        hasscon['description'].extend([desc_info.get_text()
                                       .replace('\n'," ").strip()+ ', ' + mix_info[1] + ', ' + scrap_info[1]
                                       .get_text().replace('\t','').strip()])
        
    return hasscon
```

# Property 24

```python
'https://www.property24.co.ke/property-to-rent?Page=24'
'https://www.property24.co.ke/property-for-sale?Page=2'
'https://www.property24.co.ke/developments-for-sale?Page=2'
'https://www.property24.co.ke/repossessed-property-for-sale?Page=2'
```

```python
def property24_scrapper(scrap_txt):
    property24={'prices':[],'description':[],'location':[]} #Dictionary of houses

    for a in scrap_txt:
        
        price=a.find('div', attrs={'class':'sc_listingTilePrice primaryColor'})
    
        if price == None:
            price=a.find('div', attrs={'class':'sc_listingTilePrice sc_listingTilePriceDevelopments'})
            
        property24['prices'].append(price.get_text().replace('\n'," ").strip())

        location=a.find('div', attrs={'class':'sc_listingTileAddress primaryColor'})
    
        if location == None:
            location=a.find('div', attrs={'class':'sc_listingTileAddress'})
            
        property24['location'].append(location.get_text().replace('\n'," ").strip())

        description=a.find('div', attrs={'class':'sc_listingTileArea'})
        
        if description == None:
            description=a.find('div', attrs={'class':'sc_listingTileType'})
            
        description = description.get_text().replace('\n'," ").strip()
        
        description2 = a.find('div', attrs={'class':'sc_listingTileTeaser'})
        description2 = description2.get_text().replace('\n'," ").strip()
        
        property24['description'].extend([description + ', ' + description2])
        
        

    return property24
```

```python
property24_dict={'prices':[],'description':[],'location':[]}

link_type =['developments-for-sale', 'repossessed-property-for-sale', 'property-for-sale', 'property-to-rent']
page_lmt = [5, 7, 2384, 3639]
root = 'https://www.property24.co.ke/'
pager = '?Page='

driver = wd.Chrome('./chromedriver')

for typ in link_type:
    for n in range(1,page_lmt[link_type.index(typ)]):
        link = root + typ + pager + str(n)
        driver.get(link)

        content = driver.page_source
        soup = bs(content)
        scrap_txt = soup.findAll('div',href=False, attrs={'class':'pull-left sc_listingTileContent'})

        if len(scrap_txt) == 0:
            break
            
        else:
            dict_property24 = property24_scrapper(scrap_txt)
            property24_dict['prices'].extend(dict_property24['prices'])
            property24_dict['description'].extend(dict_property24['description'])
            property24_dict['location'].extend(dict_property24['location'])
            
    
driver.quit()
property24_df = pd.DataFrame(property24_dict)
print(property24_df.sample(10))
property24_df.to_csv(path_or_buf="property24_properties.csv", index_label= "index")
```

```python
#! /usr/bin/env python
# -*- coding: utf-8 -*-
# vim:fenc=utf-8

"""

BeautifulSoup does not support XPath expression by default, so we use CSS
the expression here, but you can use https://github.com/scrapy/parsel to write
XPath to extract data as you like

"""
from __future__ import print_function
from bs4 import BeautifulSoup
import requests
from six.moves.urllib import parse

START_PAGE = "https://www.villacarekenya.com/search-properties/?location=&min_listprice=0&max_listprice=0&property_type=0&bedrooms=0&bathrooms=0&lot_area=0&floor_area=0&transaction=For+Rent&lat=&lon=&communityid=&cityid=&countyid=&subdivisionid=&locationname=&view=photo&fullscreen=lscreen=#top"

QUEUE = []

def parse_list_page(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")

    links = soup.select('a[class="page-link next-page"]')
    if links:
        next_link = links[0].attrs['href']
        next_link = url[:url.find('?')] + next_link
        QUEUE.append(
            (parse_list_page, next_link)
        )

    links = soup.select('div.col-lg-8 div.card a')
    for link in links:
        product_url = link.attrs['href']
        result = parse.urlparse(url)
        base_url = parse.urlunparse(
            (result.scheme, result.netloc, "", "", "", "")
        )
        product_url = parse.urljoin(base_url, product_url)
        QUEUE.append(
            (parse_detail_page, product_url)
        )

def parse_detail_page(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")
    print("processing " + url)

def main():
    """
    Push callback method and url to queue
    """
    QUEUE.append(
        (parse_list_page, START_PAGE)
    )

    while len(QUEUE):
        call_back, url = QUEUE.pop(0)
        call_back(url)

if __name__ == '__main__':
    main()
```

```python
def scroller_func(driver):
    SCROLL_PAUSE_TIME = 20
    
    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")
#     print(driver.current_url)

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    #     print(driver.current_url)

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            content = driver.page_source
            break
        last_height = new_height
    return content
```
